```python
import math
class Box:
    """Represents a box with dimensions length and width"""
    def __init__(self, length=0, width=0):
        self.length = length
        self.width = width
    def print_dim(self):
        print("length: " + str(self.length), "width: "+ str(self.width))    
    def __eq__(self, anotherBox):
        if(self.length == anotherBox.length):
            if(self.width == anotherBox.width):
                return True
        else: 
            return False
    def combine(self, addBox):
        self.length += addBox.length
        self.width += addBox.width
        print("Combined length: " + str(self.length), "Combined Width: " + str(self.width))
    def get_area(self):
        area = self.length * self.width
        print(area)
    def double(self):
        dbl= 2*self.length*self.width
        print(dbl)
    def get_dim(self):
        return (self.length, self.width)
    def get_hypot(self):
        hypot = math.sqrt((self.length)**2+(self.width)**2)
        print(hypot)
        
```

```python
box = Box()
box1 = Box(5,10)
box2 = Box(3,4)
box3 = Box(5,10)

box1.print_dim()
box2.print_dim()
box3.print_dim()
```
